###Ingest contructor.json file

```python
constructors_schema = 'constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING'
```

```python
constructor_df = spark.read\
    .schema(constructors_schema)\
        .json('/Volumes/workspace/f1_bronze_layer/f1_case_study/constructors.json')
```

```python
constructor_df.printSchema()
```

```python
display(constructor_df)
```

###Droping unwanted columns

```python
from pyspark.sql.functions import col,current_timestamp
```

```python
constructor_df = constructor_df.drop(col('url'))
```

```python
constructor_final_df = constructor_df.withColumnRenamed('constructorId', 'constructor_id')\
    .withColumnRenamed('constructorRef', 'constructor_ref')\
        .withColumn('ingestion_date',current_timestamp())
```

## write in parquet

```python
constructor_final_df.write.mode('overwrite').parquet('/Volumes/workspace/f1_bronze_layer/f1_case_study/constructor')
```

```python
from pyspark.sql.types import StructType,StructField,IntegerType,StringType,DateType
```

```python
name_schema = StructType(
    fields = [
        StructField('forename', StringType(), True),
        StructField('surname', StringType(), True)
    ]
)
```

```python
drivers_schema = StructType(
    fields = [
        StructField('driverId', IntegerType(), False),
        StructField('driverRef', StringType(), True),
        StructField('number', IntegerType(), True),
        StructField('code', StringType(), True),
        StructField('name', name_schema),
        StructField('dob', StringType(), True),
        StructField('nationality', StringType(), True),
        StructField('url', StringType(), True)
    ]
)
```

```python
driver_df = spark.read\
    .schema(drivers_schema)\
        .json('/Volumes/workspace/f1_bronze_layer/f1_case_study/drivers.json')
```

```python
display(driver_df)
```
