### Ingest Ciruit.csv

```python
df = spark.read.format('csv')\
    .option('inferSchema','True')\
    .option('header','true')\
    .load('/Volumes/workspace/f1_bronze_layer/f1_case_study/circuits.csv')
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType
```

```python
circuit_schema = StructType(fields = 
                            [
                                StructField('circuitId',IntegerType(),False),
                                StructField('circuitRef',StringType(),True),
                                StructField('name',StringType(),True),
                                StructField('location',StringType(),True),
                                StructField('country',StringType(),True),
                                StructField('lat',DoubleType(),True),
                                StructField('lng',DoubleType(),True),
                                StructField('alt',IntegerType(),True),
                                StructField('url',StringType(),True)
                            ]
                            )
```

```python
df = spark.read.format('csv')\
    .schema(circuit_schema)\
    .option('header','true')\
    .load('/Volumes/workspace/f1_bronze_layer/f1_case_study/circuits.csv')
```

```python
df.printSchema()
```

### Select only required column

```python
circuit_select = df.select('name','circuitId','circuitRef','lat','lng','location','country','alt')
```

```python
display(circuit_select)
```

```python
from pyspark.sql.functions import col
```

###Renaming the required column

```python
cicuit_select = df.select(col('name'),col('circuitId'),col('circuitRef'),col('lat'),col('lng'),col('location'),col('country').alias('race_country'))
```

```python
display(cicuit_select)
```

```python
circuit_select = circuit_select.withColumnRenamed('circuitID','circuit_ID')\
    .withColumnRenamed('circuitRef','circuit_ref')\
    .withColumnRenamed('lat','latitude')\
    .withColumnRenamed('lng','longitude')\
    .withColumnRenamed('alt','altitude')
```

```python
display(circuit_select)
```

### Adding a column with Ingestion date

```python
from pyspark.sql.functions import current_timestamp,lit

circuit_final_df = circuit_select.withColumn("Ingestion_date",current_timestamp())
```

```python
display(circuit_final_df)
```
